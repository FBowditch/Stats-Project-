```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
clear = 166.75
cloudy = 174.33
partly_cloudy = 166.84
rainy = 167.78
snow = 159.37
x = ['Snow', 'Clear', 'Partly Cloudy', 'Rainy', 'Cloudy']
y = [159.37, 166.75, 166.84, 167.78, 174.33]
cat = ['first', 'second', 'third', 'fourth', 'fifth']
df = pd.DataFrame(dict(x=x,y=y,cat=cat))
```

```python
graph = sns.catplot("x","y", data=df,kind="bar",palette="Blues",size=6,aspect=2,legend_out=False)
sns.set_style('dark')
plt.xlabel('Weather Type')
plt.ylabel('Mean # of Arrests')
plt.title('Weather Type VS Mean Arrests(per day)')
plt.ylim((150,175))
plt.show(graph)
```
